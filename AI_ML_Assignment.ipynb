This notebook contains both **questions and answers**.

## Q1. Explain the following with an example:
**Artificial Intelligence (AI):** Simulation of human intelligence by machines. Example: Chatbots.

**Machine Learning (ML):** Subset of AI where machines learn from data. Example: Spam detection.

**Deep Learning (DL):** Subset of ML using neural networks. Example: Face recognition.

## Q2. What is supervised learning? List examples.
Supervised learning trains a model using labeled data.

Examples:
- House price prediction
- Email spam classification
- Student result prediction

## Q3. What is unsupervised learning? List examples.
Unsupervised learning works with unlabeled data and finds hidden patterns.

Examples:
- Customer segmentation
- Market basket analysis
- Clustering

## Q4. Difference between AI, ML, DL, and DS
- AI: Intelligent machines
- ML: Learning from data
- DL: Neural network based ML
- DS: Data analysis using statistics and ML

## Q5. Differences between supervised, unsupervised, and semi-supervised learning
- Supervised: Labeled data
- Unsupervised: Unlabeled data
- Semi-supervised: Mix of labeled and unlabeled data

## Q6. Train, Test and Validation Split
- Train set: Used to train the model
- Validation set: Used to tune the model
- Test set: Used to evaluate final performance

## Q7. Unsupervised learning in anomaly detection
It helps detect unusual patterns.
Example: Fraud detection in banking.

## Q8. Common supervised and unsupervised algorithms
**Supervised:** Linear Regression, Logistic Regression, Decision Tree, Random Forest, SVM

**Unsupervised:** K-Means, Hierarchical Clustering, PCA
